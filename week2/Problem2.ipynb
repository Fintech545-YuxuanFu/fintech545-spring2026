```python
import pandas as pd
from scipy.stats import t
```

```python
def t_distribution(csv_file):
    df = pd.read_csv(csv_file)
    data = df.iloc[:, 0].dropna().values

    mu, sigma, nu = t.fit(data)
    return mu, sigma, nu

nu, mu, sigma = t_distribution("test7_2.csv")
print(mu, sigma, nu)
```
