```python
import numpy as np
import pandas as pd
```

```python
def normal_distribution(csv_file):
    df = pd.read_csv(csv_file)
    data = df.iloc[:, 0].dropna().values

    mu = np.mean(data)
    sigma = np.std(data, ddof=0)
    return mu, sigma

mu, sigma = normal_distribution(
    r"C:\Users\Zhou\FinTech-545\Week02_1\test7_1.csv"
)
print(mu, sigma)
```
